```python
import turtle
import os
import numpy as np

wn=turtle.Screen()
wn.title('Pong')
wn.bgcolor('Black')
wn.setup(width=800,height=600)

wn.tracer(0) # Stops the game from updating, so we have to do it manually

# Instructions
nep.write("Blue 'w' up 's' down")

'Initial coords always start at 0,0 center'

# Paddle A
paddle_a = turtle.Turtle()
paddle_a.speed(0) # Speed of animation, set the speed to max speed
paddle_a.shape('square')
paddle_a.shapesize(stretch_wid=5, stretch_len=1) # Stretch the size
paddle_a.color('Blue')
paddle_a.penup() # If we draw a line in Turtle, it leaves a trail of line, penup hides the line
paddle_a.goto(-350,0) # Start position

# Paddle B
paddle_b = turtle.Turtle()
paddle_b.speed(0) # Speed of animation, set the speed to max speed
paddle_b.shape('square')
paddle_b.shapesize(stretch_wid=5, stretch_len=1)
paddle_b.color('Green')
paddle_b.penup() # If we draw a line in Turtle, it leaves a trail of line, penup hides the line
paddle_b.goto(350,0) # Start position

# Paddle C
paddle_c = turtle.Turtle()
paddle_c.speed(0) # Speed of animation, set the speed to max speed
paddle_c.shape('square')
paddle_c.shapesize(stretch_wid=1, stretch_len=5)
paddle_c.color('Red')
paddle_c.penup() # If we draw a line in Turtle, it leaves a trail of line, penup hides the line
paddle_c.goto(-40,-250) # Start position

# Paddle D
paddle_d = turtle.Turtle()
paddle_d.speed(0) # Speed of animation, set the speed to max speed
paddle_d.shape('square')
paddle_d.shapesize(stretch_wid=1, stretch_len=5)
paddle_d.color('Purple')
paddle_d.penup() # If we draw a line in Turtle, it leaves a trail of line, penup hides the line
paddle_d.goto(40,-250) # Start position

# Ball
ball = turtle.Turtle()
ball.speed(0) # Speed of animation, set the speed to max speed
ball.shape('square')
ball.color('Yellow')
ball.penup() # If we draw a line in Turtle, it leaves a trail of line, penup hides the line
ball.goto(0.0,0.0) # Start position
ball.dx = 0.5
ball.dy = np.random.normal(-0.5,0.5) # Ball movement rate, the y elevation is random

#Function
def paddle_a_up():
    y = paddle_a.ycor() # Give coords
    y += 40 # Adds 20 to y-coord
    paddle_a.sety(y) # Set the new coord

def paddle_a_down():
    y = paddle_a.ycor() # Give coords
    y -= 40 # Subtracts 20 to y-coord
    paddle_a.sety(y) # Set the new coord

def paddle_b_up():
    y = paddle_b.ycor() # Give coords
    y += 40 # Adds 20 to y-coord
    paddle_b.sety(y) # Set the new coord

def paddle_b_down():
    y = paddle_b.ycor() # Give coords
    y -= 40 # Subtracts 20 to y-coord
    paddle_b.sety(y) # Set the new coord

def paddle_c_left():
    x = paddle_c.xcor()
    x -= 40
    paddle_c.setx(x)
    
def paddle_c_right():
    x = paddle_c.xcor()
    x += 40
    paddle_c.setx(x)
    
def paddle_d_left():
    x = paddle_d.xcor()
    x -= 40
    paddle_d.setx(x)
    
def paddle_d_right():
    x = paddle_d.xcor()
    x += 40
    paddle_d.setx(x)
    
#Keybinding
wn.listen() #Listen to the keyboard
wn.onkeypress(paddle_a_up,'w') # Executes the code when 'w' is pressed
wn.onkeypress(paddle_a_down,'s')
wn.onkeypress(paddle_b_up,'Up')
wn.onkeypress(paddle_b_down,'Down')
wn.onkeypress(paddle_c_left,'a')
wn.onkeypress(paddle_c_right,'d')
wn.onkeypress(paddle_d_left,'Left')
wn.onkeypress(paddle_d_right,'Right')

# Score
score_a = 0
score_b = 0
score_c = 0
score_d = 0

# Pen (Scoring system)
pen = turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()
pen.goto(0,260)
pen.write("Player A: 0   Player B: 0   Player C: 0   Player D: 0", align="center", font=("Courier",12,"normal")) # Displays inital scores

# Main Game Loop
while True :
    wn.update()
    
    # Ball Movement
    ball.setx(ball.xcor() + ball.dx)
    ball.sety(ball.ycor() + ball.dy) # From initial coords to +2
    
    # Border Check
    if ball.ycor() > 290:
        ball.sety(290)
        ball.dy *= -1 # Reverse the movement
        
    if ball.ycor() < -290:
        ball.goto(0,0)
        ball.dy *= -1
        ball.dx = np.random.normal(-0.5,0.5) # The ball moves randomly in y elevation
        score_a += 1
        score_b += 1
        pen.clear() # Resets the pen, it happens really fast
        pen.write("Player A: {}   Player B: {}   Player C: {}   Player D: {}".format(score_a, score_b, score_c, score_d), align="center", font=("Courier",12,"normal"))
        # {} means printing, when the ball goals, the score_a from the format goes to bracket
        
    if ball.xcor() > 390:
        ball.goto(0,0)
        ball.dx *= -1
        ball.dy = np.random.normal(-0.5,0.5) # The ball moves randomly in y elevation
        score_a += 1
        score_c += 1
        score_d += 1
        pen.clear() # Resets the pen, it happens really fast
        pen.write("Player A: {}   Player B: {}   Player C: {}   Player D: {}".format(score_a, score_b, score_c, score_d), align="center", font=("Courier",12,"normal"))
        # {} means printing, when the ball goals, the score_a from the format goes to bracket
        
    if ball.xcor() < -390:
        ball.goto(0,0)
        ball.dx *= -1
        ball.dy = np.random.normal(-0.5,0.5)
        score_b += 1
        score_c += 1
        score_d += 1
        pen.clear()
        pen.write("Player A: {}   Player B: {}   Player C: {}   Player D: {}".format(score_a, score_b, score_c, score_d), align="center", font=("Courier",12,"normal"))
        
    # Ball and Paddle Collision
    
    # 1st bracket (if the ball is between 340 and 350, which is the length of the paddle)
    # 2nd bracket (if the ball is between 40 and -40 which is the width of the paddle)
    
    if (ball.xcor() > 340 and ball.xcor() < 350) and (ball.ycor() < paddle_b.ycor() + 40 and ball.ycor() > paddle_b.ycor() - 40):
        ball.setx(340)
        ball.dx *= -1
        ball.dy = np.random.normal(-1,1)
        
    if (ball.xcor() < -340 and ball.xcor() > -350) and (ball.ycor() < paddle_a.ycor() + 40 and ball.ycor() > paddle_a.ycor() - 40):
        ball.setx(-340)
        ball.dx *= -1
        ball.dy = np.random.normal(-1,1)
        
    if (ball.ycor() < -240 and ball.ycor() > -250) and (ball.xcor() < paddle_c.xcor() + 40 and ball.xcor() > paddle_c.xcor() - 40):
        ball.sety(-240)
        ball.dy *= -1
        ball.dx = np.random.normal(-1,1)
        
    if (ball.ycor() < -240 and ball.ycor() > -250) and (ball.xcor() < paddle_d.xcor() + 40 and ball.xcor() > paddle_d.xcor() - 40):
        ball.sety(-240)
        ball.dy *= -1
        ball.dx = np.random.normal(-1,1)
        
    # Limiting the paddle movement within the box
    
    if paddle_b.ycor() > 260:
        paddle_b.sety(260) # Setting the coord the same as the bound makes it stop in that position
        
    if paddle_b.ycor() < -260:
        paddle_b.sety(-260)
        
    if paddle_a.ycor() > 260:
        paddle_a.sety(260)
        
    if paddle_a.ycor() < -260:
        paddle_a.sety(-260)
        
    if paddle_c.xcor() > 360:
        paddle_c.setx(360)
        
    if paddle_c.xcor() < -360:
        paddle_c.setx(-360)
        
    if paddle_d.xcor() > 360:
        paddle_d.setx(360)
        
    if paddle_d.xcor() < -360:
        paddle_d.setx(-360)
```
